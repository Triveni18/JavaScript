```python

import pandas as pd

df1 = pd.read_excel('/content/EMP.xlsx' , sheet_name="Sheet1")
df2 = pd.read_excel('/content/EMP.xlsx' , sheet_name="Sheet2")

res =pd.merge(df1 , df2 , on="EMPID", how="inner")

print(res)


```
