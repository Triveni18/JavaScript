```python
import numpy as np  # Import the NumPy library and alias it as 'np' for brevity

# Create a 1D NumPy array from a list of integers
n = np.array([9, 3, 3, 5])

# Calculate and print the arithmetic mean (average) of the array elements
# Formula: (9 + 3 + 3 + 5) / 4 = 5.0
print('Mean of the array ',n.mean())

# Calculate and print the sum of all elements in the array (20)
print('Sum of the array ', n.sum())

# Find and print the highest value in the array (9)
print('Max of the array ',n.max())

# Find and print the lowest value in the array (3)
print('Min of the array ',n.min())
```
